<h1 style="color: #2e6c80;">Problema 21.1:</h1>
Os dados da tabela vista a seguir aplicam-se à formação da ureia a partir do cianato de amônio, $NH_4CNO \rightarrow NH_{2}CONNH_{2}$. No estado inicial, 22,9 g de cianato de amônio estão dissolvidos em água suficiente para completar 1,00 $dm^{3}$ de solução. Determine a ordem da reação, a constante de velocidade e a massa de cianato de amônio remanescente depois de 300 min de reação.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Tabela de dados (data frame)
dados = {"tempo (min)": [0, 20.0, 50.0, 65.0, 150], "massa_ureia_grama (g)": [0, 7.0, 12.1, 13.8, 17.7]}
df = pd.DataFrame(dados)
df
```

<h2 style="color: #2e6c80;">1.&nbsp;</h2>A primeira coisa que iremos fazer é adicionar uma nova coluna onde calcularemos a concentração de cianato de amônio à medida que a reação progride. Para isso partimos do fato de que o problema já nos coloca que o volume da solução inicial é de $1L\; (1,00\; dm^{3})$. Portanto, a concentração calculada já estará em $mol \cdot L^{-1}$.

```python
# Adicionando coluna de concentração do cianato de amônio
df["cianato_de_amonio [mol/L]"] = np.around(((22.9 - df["massa_ureia_grama (g)"]) / (60.06)), decimals=5)
df
```

<h2 style="color: #2e6c80;">2.&nbsp;</h2>Agora, vamos construir um gráfico relacionando o tempo e a concentração de cianato de amônio e avaliar se é possível concluir alguma coisa sobre a ordem da reação.

```python
# Gráfico
fig, ax = plt.subplots(1, 1, figsize=(14, 5))
sns.lineplot(data=df, x="tempo (min)", y="cianato_de_amonio [mol/L]", marker='o')

# Títulos
plt.suptitle('Tempo x Concentração de Cianato de Amônio',
             fontweight = 'bold',
             fontsize = 25, y=1.01
             )
plt.xlabel("Tempo (min)", size=12)
plt.ylabel("Concentração []", size=12)

plt.grid()
plt.show()
```

Como temos poucos dados, o resultado não parece muito promissor. Mas é possível ver que o gráfico tem uma forma de curva, assim, já até poderíamos dizer que a ordem de nossa reação não seria zero. Assim, que tal adicionarmos mais uma coluna onde calculemos o logarítmo natural da concentração de cianato de amônio!!!

```python
# Adicionando coluna com o cálculo do logarítmo natural da concentração de cianato de amônio.
df["log [OCN]"] = np.around(np.log(df["cianato_de_amonio [mol/L]"]), decimals=5)
df
```

Depois de calculado, vamos visualizar o gráfico que relacionado o tempo de o logarítimo e avaliar o resultado.

```python
# Gráfico do tempo x logarítmo
fig, ax = plt.subplots(1, 1, figsize=(14, 5))
sns.lineplot(data=df, x="tempo (min)", y="log [OCN]", marker='o')

# Títulos
plt.suptitle('Tempo x log [OCN]',
             fontweight = 'bold',
             fontsize = 25, y=1.01
             )
plt.xlabel("Tempo (min)", size=12)
plt.ylabel("log [OCN]", size=12)

plt.grid()
plt.show()
```

Mais uma vez, parece que o gráfico apresenta uma linha reta, porém, ainda falta alguma coisa, dessa forma, não podemos ter a garantia de que a ordem da reação é 2 (dois). Portanto, façamos outra análise, calculando o inverso da concentração.

```python
df["1 / [OCN]"] = np.around(1 / df["cianato_de_amonio [mol/L]"], decimals=5)
df
```

Mais uma vez, uma visualização em gráfico pode ajudar.

```python
# Gráfico do tempo x inverso da concentração de ciananto de amônio
fig, ax = plt.subplots(1, 1, figsize=(14, 5))
sns.lineplot(data=df, x="tempo (min)", y="1 / [OCN]", marker='o')

# Títulos
plt.suptitle('Tempo x 1 / [OCN]',
             fontweight = 'bold',
             fontsize = 25, y=1.01
             )
plt.xlabel("Tempo (min)", size=12)
plt.ylabel("1 / [OCN]", size=12)

plt.grid()
plt.show()
```

Neste caso, o resultado é um forte indicador de que nossa reação é de 2ª ordem.

<h2 style="color: #2e6c80;">3.&nbsp; Regressão Linear</h2>Uma técnica que podemos empregar sobre nossos dados é a chamada regressão linear.

```python
from sklearn.linear_model import LinearRegression
```

```python
x_vlr = np.array(df["tempo (min)"]).reshape((-1, 1))
y_vlr = np.array(df["log [OCN]"])

model = LinearRegression().fit(x_vlr, y_vlr)
r_sq = model.score(x_vlr, y_vlr)
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")
```
